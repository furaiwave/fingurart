## Змінні в цьому наборі даних

###**Вік**: Вік пацієнта в роках
###**Анемія**: Зниження еритроцитів або гемоглобіну (0: Знижений або 1: Нормальний)
###**Креатинін_фосфокіназа**: рівень ферменту КФК в крові (мкг/л)
###**Діабет**: якщо у пацієнта цукровий діабет (0:Ні або 1:Так)
###**Фракція викиду**: відсоток крові, що виходить із серця при кожному скороченні (відсоток)
###**Високий кровяний тиск**: якщо у пацієнта є гіпертонія (0:Ні або 1:Так)
###**Тромбоцити**: тромбоцити в крові (кілотромбоцити/мл)
###*Cироватковий креатинін**: рівень сироваткового креатиніну в крові (мг/дл)
###**Cироватковий натрій**: рівень сироваткового натрію в крові (м-екв/л)
###**Стать**: біологічна стать пацієнта (0: жінка або 1: чоловік)
###**Куріння**: якщо пацієнт курить (0: ні або 1: так)
###**Час**: період спостереження в днях
###**Смертність**: якщо пацієнт вижив до кінця періоду спостереження (0:Ні або 1:Так)

# **Гіпотеза 1**

```python
# Створюємо датафрейм з тими даними, які більш всього коррелюють з параметром 'DEATH_EVENT'

x = df[['ejection_fraction', 'serum_creatinine', 'serum_sodium', 'time', 'age']]

# Проводимо нормалізацію даних, щоб підвищити ефективність алгоритму машинного навчання
x = (x-x.mean())/x.std()

# В окрему змінну виносимо дані, які будемо прогнозувати
y = df['DEATH_EVENT']
```

```python
x.head()
```

```python
# Розбиваємо дані на тренувальний і тестовий вибірки
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.2)
```

### **Логістична регресія** або **лоґіт-регресія**  — статистичний регресійний метод, що застосовують у випадку, коли залежна змінна є ***бінарною***, тобто може набувати тільки двох значень (0 або 1).

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

lr = LogisticRegression(max_iter=1000)
lr.fit(x_train, y_train)
p1 = lr.predict(x_test)
s1 = accuracy_score(y_test, p1)
print("Точність логістичної регресії :", "{:.2f}%".format(100 * s1))

conf_matrix = confusion_matrix(y_test, p1)
cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=[False, True])
cm_display.plot()
plt.show()
```

# **Задача 2** Порівняння методів Лог. регресії та НМ

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.src.models import Sequential
from keras.src.layers import Dense, InputLayer
from keras.src.optimizers import Adam
from keras.src.losses import MeanSquaredError
```

```python
model = Sequential([
    InputLayer(input_shape = (5,)),
    Dense(200, activation='sigmoid'),
    Dense(100, activation='sigmoid'),
    Dense(20, activation='relu'),
    Dense(1)
])

model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.003),
             loss=MeanSquaredError(),
             metrics=['accuracy'])
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
history = model.fit(x_train,y_train,  epochs=600, verbose=1)
```

```python
model.evaluate(x_test, y_test)
```

```python
#прогноз нейронної мережі
tf_history = model.predict(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

# Accuracy_score - метрика оцінки точності моделі
# Plot_confusion_matrix - матриця помилок
from sklearn.metrics import accuracy_score
from sklearn.metrics import *
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score

lr=LogisticRegression(max_iter=10000)

# fit - функція навчання моделі
lr.fit(x_train,y_train)

# predict - функція передбачення
p1=lr.predict(x_test)
s1=accuracy_score(y_test,p1)

print("Точність логістичної регресії :", "{:.2f}%".format(100*s1))

confusion_matrix = metrics.confusion_matrix(y_test, p1)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```
